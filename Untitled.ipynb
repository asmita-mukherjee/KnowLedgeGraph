```python
import matplotlib.pyplot as plt
```

### Load the data

```python
path_to_the_data = 'Data\wiki_movie_plots_deduped.csv'
```

```python
dataset = pd.read_csv(path_to_the_data)
```

```python
dataset.info()
```

```python
dataset.head()
```

### Visualize Country of origin

Top 25 country of origin

```python
dataset.groupby(['Origin/Ethnicity']).size().sort_values(ascending=True).plot.barh(y='Origin/Ethnicity')
```

## Visualize year of release

```python
list_of_movies_by_rel_year = dataset.groupby(['Release Year']).size().sort_values(ascending=True)
```

```python
fig = plt.figure(figsize =(22, 4))
plt.bar(list_of_movies_by_rel_year.keys(),list_of_movies_by_rel_year.values)

plt.show()
```

### Visualize Genre 

```python
dataset[dataset['Genre'] != 'unknown'].groupby(['Genre']).size().sort_values(ascending=True).tail(25).plot.barh(y='Genre')
```

### Visualize Plot

```python
dataset.groupby(['Title']).size().sort_values(ascending=True).tail(25).plot.barh(y='Title')
```

### Visualize Directors

```python
dataset[dataset['Director'] != 'Unknown'].groupby(['Director']).size().sort_values(ascending=True).tail(25).plot.barh(y='Director')
```

## Understanding

The most significant attributes that determines the number of movies are Country of origin and Genre
